```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def function(x):
    return (x+3)**2
```

```python
def gradientfunction(x):
    return 2*(x+3)
```

```python
def gradient_descent(function, gradient, start, learn_rate, iteration):
    func= function
    grad= gradient
    points = [start]
    iter =0
    while iter<iteration:
        prev = start
        start = start - learn_rate*grad(prev)
        points.append(start)
        iter+=1
    x_ = np.linspace(-7,5,100)
    y = func(x_)
    
    fig = plt.figure(figsize=(10,10))
    plt.plot(x_, y, 'g')
    plt.plot(points, func(np.array(points)), '-o')
    plt.show()
    
```

```python
gradient_descent(function=function, gradient= gradientfunction, start=2, learn_rate=0.2, iteration=100)
```
